```python
# Emplacement : notebooks/03_feature_engineering.ipynb
import pandas as pd
import sys
import os

# Configuration
%load_ext autoreload
%autoreload 2
sys.path.append(os.path.abspath('../src'))
from encoding import get_textual_decomposition, get_numeric_matrices

# 1. Charger le dataset
df = pd.read_csv('../data/raw/adn_dataset.csv')

# 2. Boucle k = 2, 3, 4
for k in [2, 3, 4]:
    print(f"\n--- Travail sur K = {k} ---")
    dir_path = f'../data/processed/k{k}'
    os.makedirs(dir_path, exist_ok=True)
    
    # Visualisation
    df_text = get_textual_decomposition(df, k)
    df_text.to_csv(f'{dir_path}/visualisation_k{k}.csv', index=False)
    
    # Matrices numériques (ordre : MP, PA, AT...)
    X_occ, X_bool = get_numeric_matrices(df, k)
    X_occ.to_csv(f'{dir_path}/X_occurrence_k{k}.csv', index=False)
    X_bool.to_csv(f'{dir_path}/X_boolean_k{k}.csv', index=False)
    
    print(f"   [OK] Fichiers créés. Colonnes commencent par : {X_occ.columns[:5].tolist()}")

# Labels
df['label'].to_csv('../data/processed/y.csv', index=False)
print("\nOpération terminée.")
```
