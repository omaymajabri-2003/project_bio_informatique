```python
# Dans une cellule du notebook
import pandas as pd
import sys
import os

# Configuration pour importer depuis le dossier src
sys.path.append(os.path.abspath('../src'))
from encoding import get_textual_decomposition, get_numeric_matrices

# 1. Charger le dataset brut
# Assure-toi que ce fichier contient les colonnes 'sequence' et 'label'
df = pd.read_csv('../data/processed/cleaned_data.csv')
print(f"Dataset chargé : {df.shape[0]} séquences.")

# 2. Boucle pour générer les données pour k = 2, 3 et 4
for k in [2, 3, 4]:
    print(f"\n--- Génération des caractéristiques pour K = {k} ---")
    
    # Création du dossier de destination s'il n'existe pas
    dir_path = f'../data/processed/k{k}'
    os.makedirs(dir_path, exist_ok=True)
    
    # A. Génération de la visualisation textuelle
    df_text = get_textual_decomposition(df, k)
    df_text.to_csv(f'{dir_path}/visualisation_k{k}.csv', index=False)
    
    # B. Génération des matrices numériques (X)
    X_occ, X_bool = get_numeric_matrices(df, k)
    
    # --- IMPORTANT : Ajouter la colonne label aux matrices ---
    # On utilise .values pour éviter les problèmes d'indexation
    X_occ['target'] = df['target'].values
    X_bool['target'] = df['target'].values
    
    # C. Sauvegarde des fichiers CSV finaux
    X_occ.to_csv(f'{dir_path}/X_occurrence_k{k}.csv', index=False)
    X_bool.to_csv(f'{dir_path}/X_boolean_k{k}.csv', index=False)
    
    print(f"   [SUCCÈS] Dossier k{k} mis à jour.")
    print(f"   Nombre de k-mers uniques : {X_occ.shape[1] - 1}") # -1 pour la colonne label

print("\n--- Toutes les matrices ont été générées avec succès ! ---")
```
